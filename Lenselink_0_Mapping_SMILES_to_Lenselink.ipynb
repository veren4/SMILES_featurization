In this notebook, I map canonical SMILES from ChEMBL to the dataset provided by Lenselink et al. (2017).

```python
import pandas as pd
```

```python
chembl = pd.read_table('../datasets/ChEMBL/chembl_27_chemreps.txt')
```

```python
chembl.head()
```

```python
chembl = chembl[['chembl_id', 'canonical_smiles']]
```

```python
chembl.shape
```

```python
lense = pd.read_table('../datasets/Lenselink_et_al/Dataset_files_342_MB/data/Supplementary_Information/dataset/compound_additional_physchem_features.txt',
                     low_memory=False)
```

```python
lense.head()
```

```python
# filter columns
lense = lense[['CMP_CHEMBL_ID',
               'CMP_CLASS_SMALL_MOLECULE',
               'CMP_CLASS_BIOLOGICAL',
               'CMP_CLASS_PEPTIDE',
               'CMP_CLASS_ORGANIC',
               'CMP_CLASS_INORGANIC',
               'CMP_CLASS_ACID',
               'CMP_CLASS_BASE',
               'CMP_CLASS_NEUTRAL',
               'CMP_FORMALCHARGE',
               'CMP_NUM_RINGS',
               'CMP_NUM_AROMATICRINGS',
               'CMP_NUM_CHAINS',
               'CMP_ATOMCOUNT_METALATOMS',
               'CMP_MOLECULAR_SURFACEAREA',
               'CMP_LOGP',
               'CMP_SOLUBILITY',
               'CMP_ATOMS_POSITIVE_FRAC',
               'CMP_ATOMS_NEGATIVE_FRAC',
               'CMP_MOLECULAR_WEIGHT']]
```

```python
lense.head()
```

```python
lense.shape
```

Now: create new column in lense, and add the canonical_smiles from chembl, if the chembl_id matches.

```python
# Check if the ChEMBL id is always provided in Lense
#lense.isnull().astype(int).sum()    -> there are no NULL values anywhere in lense :)
```

```python
lense = pd.merge(lense, chembl,
                how='left',
                left_on='CMP_CHEMBL_ID', right_on='chembl_id')
```

```python
lense=lense.drop(columns=['chembl_id', 'CMP_CHEMBL_ID'])
```
